```python
import pandas as pd
import glob
import re
```

```python
final_df = None
experiments_names = []

for filename in glob.glob(r"results/*"):
    df = pd.read_csv(filename)
    df = df.set_index([df.columns[0], df.columns[1]])
    
    horizons = []
    experiment_name = "".join(df.columns[0].split("_")[:-1])
    experiments_names.append(experiment_name)
    [horizons.append(x.split("_")[-1]) for x in df.columns]
    
    # print(horizons)
    # print(experiment_name)
    
    df.columns = pd.MultiIndex.from_product(([experiment_name], horizons))
    
    
    if final_df is None:
        final_df = df
    else:
        final_df = pd.concat([final_df, df], axis=1)
```

```python
order = [('Covid-19 Cases', '1'),
         ('Covid-19 Cases', '3'),
         ('Covid-19 Cases', '7'),
         ('Milk', '1'),
         ('Milk', '3'),
         ('Milk', '6'),
         ('AirlinePassengers', '1'),
         ('AirlinePassengers', '3'),
         ('AirlinePassengers', '6'),
         ('Electricity132', '1'),
         ('Electricity132', '3'),
         ('Electricity132', '7'),
         ('Electricity239', '1'),
         ('Electricity239', '3'),
         ('Electricity239', '7'),
         ('Electricity311', '1'),
         ('Electricity311', '3'),
         ('Electricity311', '7')]
```

```python
final_df = final_df[order]
```

```python
for e in experiments_names:
    df = final_df.loc[final_df.index.get_level_values('Metric') == 'MAE']
    df.index = df.index.droplevel(1)
    df = df.loc[:, df.columns.get_level_values(0) == e]
    df.columns = df.columns.droplevel(0)
    
    for h in df.columns:
        best = df.loc[df.loc[:, h] == min(df.loc[:, h])]
        print(f"{e}, {h}: {best.index[0]}")
```
